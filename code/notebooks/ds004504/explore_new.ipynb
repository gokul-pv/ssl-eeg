```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import mne
```

```python
DATASET = "ds004504"
MASTER_TABLE_PATH = f"../metadata/master_subject_table_{DATASET}.csv"

df = pd.read_csv(MASTER_TABLE_PATH)
df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


ax = sns.countplot(
    data=df, 
    x="diagnosis", 
    # hue="diagnosis", 
    # palette="viridis", 
    edgecolor="black",
    legend=False
)

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points',
                fontsize=12)
plt.title("Class distribution")
plt.show()
```

```python

plt.figure(figsize=(6,4))

ax = sns.countplot(
    data=df,
    x="diagnosis",
    hue="sex",
    # order=order
)

# add numbers on bars
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(
            f"{int(height)}",
            (p.get_x() + p.get_width()/2, height),
            ha="center",
            va="bottom",
            fontsize=11
        )

plt.title("Gender distribution per group", fontsize=14)
plt.xlabel("Group")
plt.ylabel("Count")

plt.legend(title="Sex")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.boxplot(x="diagnosis", y="age", data=df)
plt.title("Age per group")
plt.show()
```

```python
plt.figure()
sns.boxplot(x="diagnosis", y="mmse", data=df)
plt.title("MMSE per group")
plt.show()
```

```python
df["duration_min"] = df["recording_duration_sec"] / 60


# 3. Plot with styling
ax = sns.histplot(
    data=df, 
    x="duration_min", 
    bins=20, 
    kde=True,          # Adds a Kernel Density Estimate line (smooth curve)
    color="#4C72B0",   # Professional muted blue
    edgecolor="white", # Adds separation between bars
    alpha=0.8
)

# 4. Refine the layout
sns.despine()
plt.title("Distribution of Recording Durations", fontsize=12, pad=20)
plt.xlabel("Duration (Minutes)", fontsize=12)
plt.ylabel("Frequency (Count)", fontsize=12)

# 5. Optional: Add a vertical line for the Mean/Median
mean_val = df["duration_min"].mean()
plt.axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.1f}m')
plt.legend()

plt.tight_layout()
plt.show()
```
